```python
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 20 16:02:05 2021

@author: Ardhi sai kumar
"""

"""
Created on Sun Apr  4 12:46:13 2021

@author: Ardhi sai kumar
"""

#Importing lib
import smtplib
import speech_recognition as sr
import pyttsx3 
import easyimap as e

from email.message import EmailMessage    
listener = sr.Recognizer()
engine = pyttsx3.init()

#talk method
def talk(text):
    engine.say(text)
    engine.runAndWait()

# getting data from voice input method...
def get_data(x):
    try:
        with sr.Microphone() as source:
            print('listening...')
            voice = listener.listen(source, phrase_time_limit = x)
            data = listener.recognize_google(voice)
            print(data)
            return data.lower()
    except:
        pass

#method to send email.   
def send_mail(receiver, subject, content, username, password):
    server = smtplib.SMTP("smtp.gmail.com" , 587)
    server.starttls()
    server.login(username, password)
    email = EmailMessage()
    talk("can you please tell me what should I specify as your name under the receiver section")
    nick = get_data(10)
    email['From'] = nick
    email['To'] = receiver
    email['subject'] = subject
    email.set_content(content)
    server.send_message(email)

email_list = {'mark':'yuri069a@gmail.com',
              'maverick':'maverickr028@gmail.com',
              'ram':'ramu67302@gmail.com',
              'viking':'kartsuburban@gmail.com',
              'pratham':'uvpratham@gmail.com',
             }

#method to read email...
def read_mail(username, password):
    server=e.connect("imap.gmail.com", username, password)
    server.listids()
    email = server.mail(server.listids()[0])
    talk("the receiver is,")
    talk(email.from_addr)
    talk("the subject of the email is")
    talk(email.title)
    talk("the message is")
    talk(email.body)
    print(email.attachments)
    
emojis = {"laughing":"\U0001F602",
          "happy":"\U0001F603",
          "sweat":"\U0001F605",
          "smile":"\U0001F60A",
          "tasty":"\U0001F60B",
          "sleeping":"\U0001F62A",
          "crying":"\U0001F62D",
          "anger":"\U0001F620",
         }
          
def get_emoji():
            talk("I am glad you want to send some emojis please select any one of the following...")
            talk("laughing")
            talk("happy")
            talk("sweat")
            talk("smile")
            talk("tasty")
            talk("sleeping")
            talk("crying")
            talk("anger")
            emo = get_data(10)
            feed_3 = emojis[emo]
            return feed_3
            
    
    
#method to get info.
def get_info():
    talk("hello, my name is chucky, and I am your voice mail buddy...")
    talk("I will guide you throughout, the process of sending your email")
    talk("let's first get you logged in")
    
    talk("tell me your username")
    username = get_data(5)
    username = username.replace(' ', '')
    username = username + "a@gmail.com"
    
    talk("tell me your password")
    password = get_data(5)
    password = password.replace(' ','')
    
    talk("so tell me what do you want to do?")
    talk("speak send in order to compose an email or else speak read in order to read an email, simple isnt it?")
    feed = get_data(5)
    if(feed == 'send'):
        talk("tell me whom do you want to send your email")
        user_1 = get_data(5)
        receiver = email_list[user_1]
    
        talk("speak out the subject of your mail")
        subject = get_data(5)
    
        talk("Now speak out you message")
        content = get_data(25)
        talk("would you like to send some emojis?....please let me know by saying yes or no")
        feed_2 = get_data(5)
        if feed_2 == 'yes':
            emo = get_emoji()
            content = content + emo
                    
                
        send_mail(receiver, subject, content,username,password)
        talk("Mail sent buddy...")
        
    else:
        read_mail(username, password)
    
    

get_info()
```
